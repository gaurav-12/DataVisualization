#### Replicating the Seaborn Plot

#### There are 4 dimensions in the Seaborn plot. For that we need to filter the dataframe 4 times to create 4 dimensions

```python
lunch_smoke_df = tips.query("time=='Lunch' & smoker=='Yes'")

dinner_smoke_df = tips.query("time=='Dinner' & smoker=='Yes'")

lunch_nosmoke_df = tips.query("time=='Lunch' & smoker=='No'")

dinner_nosmoke_df = tips.query("time=='Dinner' & smoker=='No'")
```

#### Method 1 - Add subplots 1 by 1 to the figure

```python
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1,2,2)

sc1 = ax1.scatter(x=lunch_smoke_df.total_bill, y=lunch_smoke_df.tip, label="Yes")
sc2 = ax1.scatter(x=lunch_nosmoke_df.total_bill, y=lunch_nosmoke_df.tip, label='No')
ax1.title.set_text("time=Lunch")
ax1.set_xlabel('total_bill')
ax1.grid(alpha=0.75)
ax1.set_ylabel('tip')

sc3 = ax2.scatter(x=dinner_smoke_df.total_bill, y=dinner_smoke_df.tip, label='Yes')
sc4 = ax2.scatter(x=dinner_nosmoke_df.total_bill, y=dinner_nosmoke_df.tip, label='No')
ax2.title.set_text("time=Dinner")
ax2.set_xlabel('total_bill')
ax2.grid(alpha=0.75)
# ax2.set_ylabel('tip',fontsize=15)
# handles, labels = ax.get_legend_handles_labels()
fig.legend(handles=[sc1, sc2], loc="right", title="Smoker")


# plt.title('Iris Petal Scatterplot',fontsize=15)
# plt.legend()
plt.show()
```

#### Method 2 - Define subplots in the start and add subplots with the "axes object array"

```python
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(12, 5))
sc1 = axes[0].scatter(x=lunch_smoke_df.total_bill, y=lunch_smoke_df.tip, label="Yes")
sc2 = axes[0].scatter(x=lunch_nosmoke_df.total_bill, y=lunch_nosmoke_df.tip, label='No')
axes[0].title.set_text("time=Lunch")
axes[0].set_xlabel('total_bill')
axes[0].grid(alpha=0.75)
axes[0].set_ylabel('tip')

sc3 = axes[1].scatter(x=dinner_smoke_df.total_bill, y=dinner_smoke_df.tip, label='Yes')
sc4 = axes[1].scatter(x=dinner_nosmoke_df.total_bill, y=dinner_nosmoke_df.tip, label='No')
axes[1].title.set_text("time=Dinner")
axes[1].set_xlabel('total_bill')
axes[1].grid(alpha=0.75)
# ax2.set_ylabel('tip',fontsize=15)
# handles, labels = ax.get_legend_handles_labels()
fig.legend(handles=[sc1, sc2], loc="right", title="Smoker")
plt.show()
```

## Visualizing Time Series - Line Plot

```python
time_series = pd.DataFrame({"Datum": ['1/1/2018 0:00',
                             '1/1/2018 0:15',
                             '1/1/2018 0:30',
                             '1/1/2018 0:45',
                             '1/1/2018 1:00',
                             '1/1/2018 1:15',
                             '1/1/2018 1:30',
                             '1/1/2018 1:45 '],
                   "Value": [19.5, 19.,19.5,19.5,21,19.5,20,23]})
```

#### *Matplotlib*

```python
fig = plt.figure(figsize=(7,5))
plt.plot(time_series.Datum, time_series.Value)
plt.grid(alpha=0.75)
plt.xlabel('Datum')
plt.ylabel('Value')
plt.xticks(rotation=15)
plt.title('Time Series Plot')
plt.show()
```

#### *Seaborn*

```python
sns.lineplot(x="Datum", y="Value", data=time_series)
plt.xticks(rotation=15)
plt.show()
```

### Pie Chart

```python
tips.head()
```

```python
tips_sub_df = tips.groupby('day').sum().reset_index()[['day', 'tip','total_bill']]
```

```python
plt.figure(figsize=(5,3))
plt.pie(tips_sub_df.total_bill, labels=tips_sub_df.day, explode=(0,0,0.1,0), autopct='%1.1f%%',
        shadow=True)
plt.axis('equal')
plt.title("Daily Distribution of total bill")
plt.show()
```

### Donut Chart

```python
plt.figure(figsize=(5,3))
plt.pie(tips_sub_df.total_bill, labels=tips_sub_df.day, wedgeprops=dict(width=0.5), autopct='%1.1f%%')
plt.axis('equal')
plt.title("Daily Distribution of total bill")
plt.show()
```
